## Another method to open a file

```python
with open('new1.txt','w') as f:
    f.write("This is a new notepad text file")
```

```python
f.close()
```

```python
with open('new1.txt','r') as f:
    print(f.read())
```

```python
import csv

#csv -- comma separated values
#id ,roll_num, name
```

```python
data = [['name','email_id','phone_number'],
        ['shiavm','shivam@google.com',45652225565],
        ['komal','komal@outlook.com',45665889554]
    
]
```

```python
with open('data.csv','w') as f:
    b=csv.writer(f)
    
    for i in data:
        b.writerow(i)
```

```python
with open('data.csv','r') as f:
    read_data = csv.reader(f)
    for i in read_data:
        print(i)
```
