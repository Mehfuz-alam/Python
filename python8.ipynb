## Logical Operators
Logical operators are used to combine conditional statements

and : Return True if both statements are True

or : Returns True if one of the statements is True

not: Reverse the result, returns Flase if the result is true

```python
## And
birth_year = 1990

if birth_year > 1989 and birth_year < 1995 :
    print("You were born between 1990 and 1994")
```

```python
## Or
birth_year = 1990

if birth_year < 1980 or birth_year > 1989 :
    print("You were not born in 1980")
    
else:
    print("You were born in 1990")
```

```python
## Not
birth_year = 1990
if not birth_year == 1991:
    print("The year of birth is not 1991")
```

## Loops Statements in Python
Loops in Python are statements that allow us to perform a certain operation multiple times unless some condtitions is met.

### For Loops

For Loops is used to iterate iterables like string , tuple , list etc and perform some opertaions as shown in the flowchart below:

In for loop we need to use 'for' keyword

The for loop enables you to execute a code block multiple times

```python
# Taking an example

years = [2005,2006,2007,2008,2009,2010]

for i in years:   
    print(i)
```

```python
years = [2005,2006,2007,2008,2009,2010]

for year in years:   
    print(year)
```

## range() function
It is helpful to think of the range object of an orderred list.

This loop format will iterate overall numbers from 0 to limit - 1.

The below example prints numbers

```python
for i in range(5):
    print(i)
```

```python
# Take a range() function 

print(range(5))
print(range(10))
```

```python
years = [2005,2006,2007,2008,2009,2010]

for i in range(len(years)):   
    print(years[i])
```

```python
for i in range(2,10,2):
    print(i)
```

```python
for i in range(2,12,3):
    print(i)
```

```python
# Changing the elements in the list
lang = ["Java","JavaScript", "C","C++","PHP"]

for i in range(len(lang)):
    print('Before language',i ,'is',lang[i])
    lang[i] = 'Python'
    print('After language',i ,'is',lang[i])
```
