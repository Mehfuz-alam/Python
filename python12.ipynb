## Generator Functions
Python generators are easy way of creating iterators. It generates values one at a time from a given sequence instead of returning the entire sequence at once.

In a genrator function, a 'yield' statement is used rather than a return statement.

It is a special type of function which retuns an iterator object.

When the generator function is called , it returns an iterator and values can be generated by using the 'next()' function.

The difference between yield and return is that once yield returns a value the function is paused and the control is transferred to the caller. Local variables and their states are rememberred btw successive calls. In the case of the return statement value is returned and the execution of the function is terminated.

Methods like iter() and next() are implemented automaically in generator function.

Where we need to use generator function in data science ----Large Datasets

When working with the massive datasets that may not fit into the memory, using a generator allows you to process data in chunks, one pience at a time. This helps in avoiding memory overflow.

```python
# Normal function
def normal_function():
    return [1,2,3,4,5]

# Generator function
def generator_function():
    for i in range(1,6):
        yield i
        


#using normal function
result_normal = normal_function()
print(result_normal)

#uisng generator function 
result_generator = generator_function()
print(next(result_generator)) #1
print(next(result_generator))  #2

```

```python
print(next(result_generator))  #3
print(next(result_generator))  #4
print(next(result_generator))   #5
```

```python
# simple generator function that will generate numbers from 1 to 5
def mygen():
    n = 1
    yield n
    
    n += 1
    yield n
    
    n += 1
    yield n
    
    n += 1
    yield n
    
    n += 1
    yield n
    
mygen1 = mygen()

print(next(mygen1))
print(next(mygen1))
print(next(mygen1))
print(next(mygen1))
print(next(mygen1))   #Function will terminate here as all 5 values have been returned
print(next(mygen1))   #As function is already terminated , StopIteration is raised
```

```python
def my_gen():
    for i in range(1,20):
        yield i
        
mygen1 = my_gen()

for i in mygen1:
    print(i)
```

```python
def my_gen():
    for i in range(1,20):
            if i%2 == 0:
                yield i
        
mygen1 = my_gen()

for i in mygen1:
    print(i)
```

```python
def test_fib(n):
    a,b = 0,1
    for i in range(n):
        yield a
        a,b=b,a+b
        
for i in test_fib(22):
    print(i)
```

```python
def test_fib1():
    a,b = 0,1
    while True:
        yield a 
        a,b = b, a+b
```

```python
def count_test(n):
    count = 1
    while count <=n :
        yield count
        count = count + 1
```

## lambda Function
These are small anonymous functions in python, which can take any number of arguments but return only 1 expression

```python
#consider the funtion which multiplies 2 numbers

#By using a normal function

def mul(a,b):
    return(a*b)
print(mul(3,2))
```

```python
mul=lambda a,b : a*b
print(mul(3,2))
```

```python
n = 3
p = 2
def test(n,p):
    return n**p
test(3,2)
```

```python
a=lambda m,n:m*n
```

```python
a(3,2)
```
